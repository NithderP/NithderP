#Data Statistics

```python
# Define a function to generate statistics for each column
def generate_stats(column):
    dtype = str(column.dtype)
    stats = {
        'Field Type': dtype,
        'Null Values': column.isnull().sum(),
        'Unique Values': column.nunique(),
        'Count': column.count(),
    }
    if pd.api.types.is_numeric_dtype(column):
        stats['Mean'] = column.mean()
        stats['Std Deviation'] = column.std()
        stats['Min'] = column.min()
        stats['25% (1st Quartile)'] = column.quantile(0.25)
        stats['Median (2nd Quartile)'] = column.median()
        stats['75% (3rd Quartile)'] = column.quantile(0.75)
        stats['Max'] = column.max()
    return pd.Series(stats)
```

```python
# Generate statistics for each data type
numerical_columns = df.select_dtypes(include=['number'])
ordinal_columns = df[['WklyStudyHours', 'ParentEduc', 'PracticeSport']]
categorical_columns = df[['ParentMaritalStatus', 'Gender','EthnicGroup', 'LunchType', 'TestPrep', 'IsFirstChild', 'TransportMeans']]

numerical_stats = numerical_columns.apply(generate_stats)
ordinal_stats = ordinal_columns.apply(generate_stats)
categorical_stats = categorical_columns.apply(generate_stats)

# Concatenate the statistics for all data types into a single DataFrame
all_stats = pd.concat([numerical_stats, ordinal_stats, categorical_stats], axis=1).T

# Create a DataFrame with proper column names
all_stats_df = pd.DataFrame(all_stats, columns=all_stats.columns)

all_stats_df
```

#Find outliers

```python
# Function to detect unique outliers using quantiles and return them as a list

def find_unique_outliers(column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)][column]
    return list(set(outliers))  # Convert to a list with unique values

outliers_dict = {}

# Find unique outliers for all numerical columns and store them in a dictionary

attributes_with_outliers = df.select_dtypes(include=['number']).columns
for attribute in attributes_with_outliers:
    column_outliers = find_unique_outliers(attribute)
    if column_outliers:
        outliers_dict[attribute] = column_outliers

# Display the dictionary of attributes with unique outliers
print("Attributes with unique outliers and their values:")
for attribute, unique_outliers in outliers_dict.items():
    print(attribute, ":", unique_outliers)
```
